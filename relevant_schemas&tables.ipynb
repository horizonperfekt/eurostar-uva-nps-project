```python
import delta_sharing
CREDENTIAL = "config.share" 
client = delta_sharing.SharingClient(CREDENTIAL)
SHARE_NAME = 'amsterdam_students_share' 
directory = 'Data'
```

```python
directory = 'Data/Equinox'
equinox_schema = "equinox_reporting" 
equinox_tables = [ "dim_priority"
                  #"fct_work_reports", 
                  #"dim_stock_type", 
                  #"dim_task", 
                  #"fct_stock_issue", 
                  #"dim_fault_code", 
                  #"dim_campaign"
                ] 
for table in equinox_tables: 
    table_name = table 
    equinox_url = f"{CREDENTIAL}#{SHARE_NAME}.{equinox_schema}.{table}" 
    table = delta_sharing.load_as_pandas(equinox_url) 
    table.to_parquet(f"{directory}/{table_name}.parquet", index=False)
```
