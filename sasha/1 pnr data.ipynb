```python
# What passenger types exist
print(df_booking['passenger_type'].value_counts())

# Group size distribution (passengers per PNR)
group_sizes = df_booking.groupby('pnr_identifier')['tcn'].nunique()
print(group_sizes.describe())
print(group_sizes.value_counts().head(10))

# Do groups tend to cluster in the same coach?
df_booking.groupby('pnr_identifier').agg(
    group_size=('tcn', 'nunique'),
    coaches=('coach_number', 'nunique')
).query('group_size > 1').head(10)
```
